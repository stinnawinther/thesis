```python
#Install packages:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches

import requests
import math
import csv
```

```python
#read metadata from txt to df:
df = pd.read_csv('../Data from datasets/COVID-19_image_data_collection_metadata.txt')
```

```python
df
```

```python
CXR_df = df.groupby(['modality','folder'])[['patientid']].count().reset_index()
```

```python
CXR_df=df[df["modality"]=='X-ray']
```

```python
CXR_df
```

```python
location_df = CXR_df.groupby('location')[['patientid']].count().reset_index()
```

```python
location_df
```

```python
CXR_df['location']
```

```python
#select the country of each location: 
CXR_df['country'] = CXR_df["location"].str.rsplit(",", 1).str[-1]
```

```python
CXR_df['country'] = CXR_df['country'].str.lower()
```

```python
CXR_df['country']
```

```python
#drop missing values:
CXR_df['country'].dropna()
```

```python
countries = CXR_df.groupby('country')[['patientid']].count().reset_index()
```

```python
countries
```

```python
#save countries as csv:
countries.to_csv("countries.csv")
```
