```python
#Install packages:
import numpy as np
import matplotlib as plt
import pandas as pd
import matplotlib.patches as mpatches

import requests
import math
import csv
```

```python
#read identified datasets from google dataset:
datasets_Google = pd.read_csv('../thesis/GoogleDataset_Datasets.csv')
```

```python
#read identified datasets from OpenAlex:
datasets_OpenAlex = pd.read_csv('../thesis/OpenAlex_Datasets.csv')
```

```python
all_datasets_df = pd.concat([datasets_Google,datasets_OpenAlex]).drop_duplicates().reset_index(drop=True)
```

```python
#The TBX dataset is included as it consists of both TBX - Dataset A and TBX - Dataset B.
#Therefore, TBX (index 46) is removed:
all_datasets_df = all_datasets_df.drop(index=45)
```

```python
# The RSNA and RSNA/RICORD is the same. Therefore RSNA is removed (index=44)
all_datasets_df = all_datasets_df.drop(index=44)
```
