```python
#required:
urls = []
titles = []
descriptions=[]
```

```python
from googlesearch import search
search_results = search("chest radiology", advanced=True)
```

```python
for result in search_results:
    url = result.url
    title=result.title
    description = result.description
    
    urls.append(url.encode('ascii', 'ignore').decode('ascii'))
    titles.append(title.encode('ascii', 'ignore').decode('ascii'))
    descriptions.append(description.encode('ascii', 'ignore').decode('ascii'))
```

```python
df = pd.DataFrame({'url':urls, 'title':titles, 'description':descriptions})
```
